```python
import numpy as np
from tensorflow.keras.models  import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
import numpy as np

X = []
y = []
for i in range(1, 10):
    X.append([i, i+1, i+2])
    y.append(i+3)

X = np.array(X).reshape((len(X), 3, 1))
y = np.array(y)
```

```python
model = Sequential()
model.add(SimpleRNN(units=10, input_shape=(3, 1), activation='relu'))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
model.fit(X, y, epochs=300, verbose=1)
```

```python
test_input = np.array([[10, 11, 12]]).reshape((1, 3, 1))
predicted = model.predict(test_input, verbose=0)

print(f"Input Sequence: [10, 11, 12]")
print(f"Predicted Next Number: {round(float(predicted[0][0]), 2)}")
```
